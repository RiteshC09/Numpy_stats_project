# NumPy Project

```python
data_x = [10, 15, 2, 43, 89, 49, 21]
data_y = [28, 33, 79, 67, 39, 19, 72]
```

```python
import numpy as np
import statistics

class WizbiStats:
    def __init__(self, data_x, data_y):
        self.x = data_x
        self.y = data_y

    def length(self):
        len_x = len(self.x)
        len_y = len(self.y)
        return f'length of X:{len_x}, \nlength of Y:{len_y}'
    
    def sum(self):
        sum_x = np.sum(self.x)
        sum_y = np.sum(self.y)
        return f'sum of dataset X: {sum_x}, \nsum of dataset Y: {sum_y}'

    def mean(self):
        mean_x = round(np.mean(self.x), 2)
        mean_y = round(np.mean(self.y), 2)
        return f'mean of data_x: {mean_x}, \nmean of data_y: {mean_y}'

    def median(self):
        median_x = np.median(self.x)
        median_y = np.median(self.y)
        return f'mean of dataset X: {median_x}, \nmean of dataset Y: {median_y}'

    def mode(self):
        mode_x = statistics.mode(self.x)
        mode_y = statistics.mode(self.y)
        return f'mode of dataset X: {mode_x}, \nmode of dataset Y:{mode_y}'


    def varience(self):
        var_x = round(np.var(self.x), 2)
        var_y = round(np.var(self.y), 2)
        return f'varience of dataset X: {var_x}, \nvarience of dataset Y: {var_y}'


    def std_deviation(self):
        std_x = round(np.std(self.x), 2)
        std_y = round(np.std(self.y), 2)
        return f'standard deviation of dataset X: {std_x}, \nstandard deviantion of dataset Y: {std_y}'

    
    def correlation(self):
        corr_x_y = np.corrcoef(self.x, self.y)
        return f' correlation between dataset X and dataset Y is: {round(corr_x_y[0][1], 3)}'

    
    def data_range(self):
        range_x = max(self.x) - min(self.x)
        range_y = max(self.y) - min(self.y)
        return f'Range of dataset X: {range_x}, \nRange of dataset Y: {range_y}'
        

    
    def quartiles(self):
        q1x = np.percentile(self.x , 25)
        q2x = np.percentile(self.x , 50)
        q3x = np.percentile(self.x , 75)

        q1y = np.percentile(self.y , 25)
        q2y = np.percentile(self.y , 50)
        q3y = np.percentile(self.y , 75)

        return q1x, q2x, q3x, q1y, q2y,q3y


    def interquartile_range(self):
        q = self.quartiles()
        iqr_x = q[2] - q[0]
        iqr_y = q[5] - q[3]
        return f'Interquartile range of dataset X: {iqr_x}, \nInterquartile Range of dataset Y: {iqr_y}'

    
    def confidence_interval(self):
        mean_x, std_x, len_x = np.mean(self.x), np.std(self.x), len(self.x)
        mean_y, std_y, len_y = np.mean(self.y), np.std(self.y), len(self.y)

        ci_x = ( 
            round(mean_x - (1.96 * (std_x / (np.sqrt(len_x)))), 2),
            round(mean_x + (1.96 * (std_x / (np.sqrt(len_x)))), 2),
        )

        ci_y = ( 
            round(mean_y - (1.96 * (std_y / (np.sqrt(len_y)))), 2),
            round(mean_y + (1.96 * (std_y / (np.sqrt(len_y)))), 2),
        )
        return f' confidence_interval of daatset X: {ci_x}, confidence_interval of dataset Y: {ci_y}'
```

```python
stats_analysis = WizbiStats(data_x, data_y)
```
